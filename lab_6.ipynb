```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

mean, std_dev, num_points = 30, 7, 1000
data = np.random.normal(mean, std_dev, num_points)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram and PDF')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
sns.kdeplot(data, bw_adjust=0.5)
plt.title('Kernel Density Estimate')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean, std_dev = 50, 10
x = np.linspace(0, 100, 1000)
pdf = norm.pdf(x, mean, std_dev)

prob_greater_than_30 = 1 - norm.cdf(30, mean, std_dev)
prob_less_than_30 = norm.cdf(30, mean, std_dev)
prob_between_30_and_50 = norm.cdf(50, mean, std_dev) - norm.cdf(30, mean, std_dev)

plt.plot(x, pdf, label='Normal Distribution', color='blue')
plt.fill_between(x, pdf, where=(x >= 30), color='green', alpha=0.5, label='P(X ≥ 30 bpm)')
plt.fill_between(x, pdf, where=(x <= 30), color='red', alpha=0.5, label='P(X ≤ 30 bpm)')
plt.fill_between(x, pdf, where=(x >= 30) & (x <= 50), color='orange', alpha=0.5, label='P(30 < X < 50 bpm)')

plt.title('Resting Heart Rate Distribution')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

print(f"P(X ≥ 30 bpm): {prob_greater_than_30:.4f}")
print(f"P(X ≤ 30 bpm): {prob_less_than_30:.4f}")
print(f"P(30 < X < 50 bpm): {prob_between_30_and_50:.4f}")
```
