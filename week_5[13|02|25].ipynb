```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
daily_calls = np.random.randint(0, 8, N)

call_values, call_counts = np.unique(daily_calls, return_counts=True)
pmf = call_counts / N
cdf = np.cumsum(pmf)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(call_values, pmf)
plt.xlabel("Number of Calls")
plt.ylabel("Probability")
plt.title("PMF of Daily Calls")

plt.subplot(1, 2, 2)
plt.plot(call_values, cdf, marker='o', linestyle='-')
plt.xlabel("Number of Calls")
plt.ylabel("Cumulative Probability")
plt.title("CDF of Daily Calls")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import binom

n = 100
p = 0.25
k = 20

prob_exact_20 = binom.pmf(k, n, p)
prob_at_least_30 = 1 - binom.cdf(29, n, p)
expected_surgery = n * p

print(f"Probability of exactly 20 patients needing surgery: {prob_exact_20:.4f}")
print(f"Probability of at least 30 patients needing surgery: {prob_at_least_30:.4f}")
print(f"Expected number of patients requiring surgery: {expected_surgery:.1f}")
```
